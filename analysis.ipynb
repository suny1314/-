```python
import pandas as pd
import matplotlib.pyplot as plt

# Assuming 'result' is the DataFrame containing the data

# Calculate the average savings per used time for each appliance type
result['average_savings_per_year'] = result['savings(pences)'] / result['years'] /100
result_filtered = result[result['years'] != 0]
average_savings_per_appliance = result_filtered.groupby('appliancetype')['average_savings_per_year'].mean()

# Sort the data by average savings per time in descending order for better visualization
average_savings_per_appliance = average_savings_per_appliance.sort_values(ascending=False)

# Create the bar plot
plt.figure(figsize=(10, 6))
average_savings_per_appliance.plot(kind='bar', color='skyblue')
plt.xlabel('Appliance Type', fontsize=15)
plt.ylabel('Average Savings per year (pounds)', fontsize=15)
plt.title('Average Savings per year of Different Appliance Types')
plt.xticks(rotation=45, ha='right', fontsize=14)
plt.yticks(fontsize=14)  # Increase fontsize for y-axis tick labels

plt.grid(axis='y')
plt.tight_layout()
plt.show()
```
